```python
class Contador:
    def __init__(self, Limite):
        self.Limite = Limite
        self.atual = 1
        
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.atual > self.Limite:
            raise StopIteration
        else:
            valor = self.atual
            self.atual += 1
            return valor
```

```python
contador = Contador(5)
for numero in contador:
    print(numero)
```
