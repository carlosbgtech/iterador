```python
import itertools
```

```python
# Exemplo 1 - Range()
for i in range(5):
    print(i)
```

```python
# Exemplo 2 - Iterar Lista
frutas = ['Maçã', 'Banana', 'Laranja']
for fruta in frutas:
    print(fruta)
```

```python
# Exemplo 3 - Iterar Dicionário
pessoa = {'nome': 'Carlos', 'idade': 43,}
for chave in pessoa:
    print(chave, pessoa[chave])
```

```python
# Exemplo 4 - Enumerate()
for indice, fruta in enumerate(frutas):
    print(indice, fruta)
```

```python
# Exemplo 5 - Zip()
nomes = ['Alice', 'Bob', 'Charlie']
idades = [25, 30, 35]
for nome, idade in zip(nomes, idades):
    print(nome, idade)
```

```python
# Exemplo 6 - Itertools
for i in itertools.count(start=10, step=2):
    if i > 20:
        break
    print(i)
```
