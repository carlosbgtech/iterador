Filter
- É usada para filtrar elementos de um iterável com base em uma função que pode ser True ou False.
- o objetivo é permitir apenas os elementos que satisfazem uma condição.

```python
def e_par(x):
    return x % 2 == 0
```

```python
numeros = [1, 2, 3, 4, 5, 6]
```

```python
numeros_pares = filter(e_par, numeros)
list(numeros_pares)
```

```python
produtos = [
    {"nome": "Produto A", "preco": 10.0, "em_estoque": True},
    {"nome": "Produto B", "preco": 20.0, "em_estoque": False},
    {"nome": "Produto C", "preco": 30.0, "em_estoque": True},
    {"nome": "Produto D", "preco": 45.0, "em_estoque": False},
]
```

```python
def produto_em_estoque(produto):
    return produto["em_estoque"]
```

```python
produtos_disponiveis = filter(produto_em_estoque, produtos)
list(produtos_disponiveis)
```
