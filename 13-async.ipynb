Async
- Permite que você escreva código que pode ser executado de forma não bloqueante.
- Facilita a execução de múltiplas operações simultâneas.

```python
import asyncio
import aiohttp
```

```python
async def tarefa(nome, tempo):
   print(f"Tarefa {nome} inciada.")
   await asyncio.sleep(tempo)
   print(f"Tarefa {nome} concluída.")
```

```python
async def main():
    await asyncio.gather(
        tarefa("A", 2),
        tarefa("B", 1)
    )
```

```python
try:
    asyncio.get_running_loop()
except RuntimeError:
    asyncio.run(main())
```

```python
asyncio.ensure_future(main())
```

```python
async def buscar_usuario(session, user_id):
   async with session.get(f'https://jsonplaceholder.typicode.com/users/{user_id}') as resposta:
       return await resposta.json()
```

```python
async def main():
     async with aiohttp.ClientSession() as session:
        usuario_ids = [1, 2, 3, 4, 5]
        tarefas = [buscar_usuario(session, user_id) for user_id in usuario_ids]
        
        # Aguarda todas as tarefas serem concluídas
        usuarios = await asyncio.gather(*tarefas)
        for usuario in usuarios:
            print(usuario)
```

```python
try:
    asyncio.get_running_loop()
except RuntimeError:
    asyncio.run(main())
```

```python
asyncio.ensure_future(main())
```
